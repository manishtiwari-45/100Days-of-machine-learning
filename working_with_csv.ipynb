## Opening a csv file from an URL

```python
import requests
from io import StringIO

url = ""
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mx)"}
req = requests.get(url,headers = headers)
data = StringIO(req.text)

pd.read_csv(data)
```
